```python
# 1)

import pandas as pd
import numpy as np
import itertools


data = pd.read_csv("market_basket.txt",sep="\t")
data.sort_values(by=['ID'], inplace=True)
```

```python
# 4)

row = data["ID"].unique()
column = data["Product"].sort_values().unique()
matrix = np.zeros((len(row), len(column)))
compteur = 0

for i in row:
    data_frame = data[data["ID"] == i]
    a = 0
    for j in column:
        if data_frame["Product"].str.contains(j).any():
            matrix[compteur][a] = 1
        a = a + 1
    compteur = compteur + 1

data_frame = pd.DataFrame(matrix, index=row, columns=column)
data_frame
```

```python
# 5)

data_frame = pd.crosstab(data["ID"], data["Product"])
data_frame
```

```python
# 6)

data_frame.iloc[:3, :30]
```

```python
# 7)

max = 2


def extraire_itemset(data_frame, n):
    return list(itertools.combinations(data_frame, n))


def support(df, subsets, n):
    tab = []
    k = 0
    for i in subsets:
        subset = df[list(i)]
        w = subset[subset.sum(axis="columns") == n].count()
        tab.append(w[0])
        k = k + 1
    return tab


support_minimal = 0.025 * len(row)

for i in range(1, max + 1):
    itemset = extraire_itemset(column, i)
    Support = support(data_frame, itemset, i)
    ff = pd.DataFrame(itemset)
    sup = pd.DataFrame(Support)
    sup.columns = ["Support"] 
    frame = [ff, sup] 
    F = pd.concat(frame, axis=1)
    C = F[F["Support"] > support_minimal]
    print("C", i)
    print(C)


def regle(item):

    it = item[:-1]
    for i in range(1, max):
        List = extraire_itemset(it, i)
        permis = List[:-i]
        cc = List[i:]
        support_1 = support(data_frame, permis, i)
        supports = support(data_frame, [it], max)
    return (permis, cc, np.array(supports) / np.array(support_1))


resultats = []
compteur = 0

for i in C.index:
    compteur = compteur + 1
    ligne = list(C.loc[i, :])
    premis, Ccl, Confidence = regle(ligne)
    print("régle :", premis, " : ", Ccl, " confidence:", Confidence)

from mlxtend.frequent_patterns import apriori, association_rules

items_frequents = apriori(data_frame, min_support=0.025, use_colnames=True)

regles = association_rules(items_frequents, metric="lift", min_threshold=1)
regles = regles.sort_values(["confidence", "lift"], ascending=[False, False])

print(regles.head())
```

```python
# 9)

def is_inclus(_subset, _set):
    if len(_subset) > len(_set):
        return False
    for i in _subset:
        if not (i in _set):
            return False
    return True
```

```python
# 10)

régles = [
    list(a) + list(c) for a, c in zip(regles["antecedents"], regles["consequents"])
]
for i in régles:
    if is_inclus(["Aspirin"], i):
        print(i)
```
